This notebook is to look on other python packages that can read rds

I'll test packages on two objects:

1. mtcars.rds - data.frame
2. GSE268542_HuLu2023.RDS - Seurat object

see tutorial.ipynb for data download

# rdata

```python
import rdata
```

```python
%%time
converted = rdata.read_rds('mtcars.rds')
print(converted)
```

```python
%%time
converted = rdata.read_rds('GSE268542_HuLu2023.RDS')
print(converted)
```

# pyreadr

```python
import pyreadr
```

```python
%%time
result = pyreadr.read_r('mtcars.rds')
print(result)
```

```python
%%time
result = pyreadr.read_r('GSE268542_HuLu2023.RDS')
print(result)
```

# rds2py

```python
from rds2py import read_rds
```

```python
%%time
r_obj = read_rds("mtcars.rds")
r_obj
```

```python
%%time
r_obj = read_rds("GSE268542_HuLu2023.RDS")
r_obj
```
